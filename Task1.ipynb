```python
plt.figure(figsize=(10, 6))
plt.bar(x.State,x.Population, color='skyblue')
plt.xlabel('States')
plt.ylabel('Population')
plt.title('Population by State')
plt.xticks(rotation=90)  
plt.tight_layout()  
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(x.State,x.Males, color='skyblue')
plt.xlabel('States')
plt.ylabel('male population')
plt.title('male population by State')
plt.xticks(rotation=90)  
plt.tight_layout()  
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(x.State,x.Females, color='skyblue')
plt.xlabel('States')
plt.ylabel('female population')
plt.title('female population by State')
plt.xticks(rotation=90)  
plt.tight_layout()  
plt.show()
```
